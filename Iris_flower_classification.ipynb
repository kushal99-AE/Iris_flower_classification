## Model Developement 

```python
#Importing required librarires
import sklearn
from sklearn.model_selection import train_test_split
```

Assigning two different arrays - one for Unqiue features of data(which are first four columns) - Input  and species column as - output

```python
x = iris_dataset.iloc[:,:4]
y = iris_dataset.iloc[:,4]
```

```python
x
```

```python
y
```

```python
#Splitting the data into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 0)
```

```python
#View the test and train data shape
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Create the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy = accuracy_score(y_test,y_pred)*100
```
